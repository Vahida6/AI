**TASK**

**A. Build a sequential model using Keras on top of this Diabetes dataset to find out if the
patient has diabetes or not, using ‘Pregnancies’, ‘Glucose’ & ‘BloodPressure’ as
independent columns.
a. This model should have 1 hidden layer with 8 nodes
b. Use Stochastic Gradient as the optimization algorithm
c. Fit the model, with number of epochs to be 100 and batch size to be 10**

```python
df.info()
```

```python
df.isnull().sum()
```

NO NULL Values so will model directly

```python
x=df.iloc[:,:3]
x
```

```python
y=df['Outcome']
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_scaled=sc.fit_transform(x)
x_scaled[1:10]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2)
```

```python
x_train.shape
```

```python
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential


# create model
model = Sequential()
model.add(Dense(8, input_shape=(3,), activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
epochs = 100
learning_rate = 0.1
decay_rate = learning_rate / epochs
momentum = 0.8
sgd = SGD(learning_rate=learning_rate, momentum=momentum, decay=decay_rate, nesterov=False)
model.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
# Fit the model
model.fit(x_train, y_train, validation_split=0.33, epochs=epochs, batch_size=10, verbose=2)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
import numpy as np
y_pred = model.predict(x_test)
y_pred_classes = [np.argmax(element) for element in y_pred]

print("Classification Report: \n", classification_report(y_test, y_pred_classes))
```

```python
predictors=df.iloc[:,:8]
predictors
```

```python
target=df['Outcome']
target
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
predictors_scaled=sc.fit_transform(x)
predictors_scaled[1:10]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(predictors_scaled,target,test_size=0.2)
```

```python
x_train.shape
```

** ‘adam’ as the optimization algorithm **

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(16, activation='relu', input_shape=[8]),
    layers.Dense(16, activation='relu'),
    layers.Dense(16, activation='relu'),
    layers.Dense(1),
])
model.compile(
    optimizer='adam',
    loss='mse',
     metrics=['accuracy'])

model.fit(x_train, y_train, validation_split=0.33, epochs=150, batch_size=10, verbose=2)

from sklearn.metrics import confusion_matrix , classification_report
import numpy as np
y_pred = model.predict(x_test)
y_pred_classes = [np.argmax(element) for element in y_pred]

print("Classification Report: \n", classification_report(y_test, y_pred_classes))
```
